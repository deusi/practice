## LC #190
Status: Acceptable, but runtime & space are slow, so there's a more elegant solution

```python
class Solution:
  # Time: 19 m
  # Runtime Complexity: O(32) since we traverse at most 32 elements
  # Space Complexity O(32) since we create a new digit
  def reverseBits(self, n: int) -> int:
    reversed = ['0'] * 34
    reversed[1] = 'b'
    orig = str(bin(n))
    for i in range(len(orig) - 2):
      reversed[len(orig) - 1 - i] = str(orig[2 + i])
    return int("".join(reversed), 2)

def main():
  sol = Solution()
  print(sol.reverseBits(43261596))
  print(sol.reverseBits(964176192))
  print(sol.reverseBits(2147483644))

main()
```

## LC #401
Status: Solved, but not the most optimal + lengthy

```python
class Solution:
    def readBinaryWatchRec(self, hours, minutes, turnedOn, curHours, curMinutes, times, curI, curJ):
            if curHours > 11 or curMinutes > 59 or turnedOn < 0:
                return
            if turnedOn == 0:
                if curHours < 12 and curMinutes < 60:
                    if curMinutes < 10:
                        timeStr = str(curHours) + ':0' + str(curMinutes)
                    else:
                        timeStr = str(curHours) + ':' + str(curMinutes)
                    times.add(timeStr)
            else:
                for i in range(curI, len(hours)):
                    self.readBinaryWatchRec(hours, minutes, turnedOn - 1, curHours + hours[i], curMinutes, times, i + 1, curJ)
                for j in range(curJ, len(minutes)):
                    self.readBinaryWatchRec(hours, minutes, turnedOn - 1, curHours, curMinutes + minutes[j], times, curI, j + 1)

    # time: 35 m
    # Runtime Complexity: O(turnedOn * minutes * hours)
    # Space Complexity: O(minutes * hours)
    def readBinaryWatch(self, turnedOn: int) -> List[str]:
        hours = [8,4,2,1]
        minutes = [32,16,8,4,2,1]
        times = set()

        self.readBinaryWatchRec(hours, minutes, turnedOn, 0, 0, times, 0, 0)
        return list(times)

```

## LC #67

Status: Solved, but lengthy and ugly. Recommend revisiting

```python
class Solution:
  # 49 m
  # Time Complexity: max(len(a), len(b))
  # Space Complexity: max(len(a), len(b))
  def addBinary(self, a: str, b: str) -> str:
    addition = []
    shorter, longer = (a, b) if len(a) <= len(b) else (b, a)
    carry = 0
    for i in range(min(len(a), len(b))):
      if a[-1-i] != b[-1-i]:
        if not carry:
          addition.append('1')
        else:
          addition.append('0')
      else:
        addition.append(str(carry))
        carry = 0
        if a[-1-i] == '1':
          carry = 1

    for j in range(len(shorter), len(longer)):
      if longer[-1-j] == '1' and carry:
        addition.append('0')
      elif longer[-1-j] == '0' and carry:
        addition.append(str(carry))
        carry = 0
      else:
        addition.append(longer[-1-j])

    if carry:
      addition.append(str(carry))


    return ''.join(addition[::-1])


def main():
  sol = Solution()
  print(sol.addBinary("11", "1"))
  print(sol.addBinary("11", "11"))

  print(sol.addBinary("1010", "1011"))

  print(sol.addBinary("1", "1"))
  print(sol.addBinary("111", "111"))
  print(sol.addBinary("100", "111"))

  print(sol.addBinary("100", "110010"))

  print(sol.addBinary("110010", "10111"))

main()
```
