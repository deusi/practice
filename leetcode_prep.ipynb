## LC #3666
Status: Solved, but the solution looked up

```python
class Solution:
    # Time: 14 mins (after looking up solution guide)
    # Runtime: O(n)
    # Space: O(1)
    def minOperations(self, s: str, k: int) -> int:
        n, zeros, ones = len(s), s.count('0'), s.count('1')
        if n == ones:
            return 0
        for op in range(1, n + 1):
            flips = op * k
            if (flips - zeros) % 2 == 1:
                continue
            if op % 2 == 0 and zeros <= flips <= zeros * (op - 1) + ones * op:
                return op
            elif op % 2 == 1 and zeros <= flips <= zeros * op + ones * (op - 1):
                return op
        return -1


def main():
    sol = Solution()
    print(sol.minOperations("110", 1))
    print(sol.minOperations("0101", 3))
    print(sol.minOperations("101", 2))


main()
```

## LC #1356
Status: solved

```python
from typing import List

class Solution:
    # Time: 23 m
    # Runtime: O(nlogn)
    # Complexity: O(n)
    def sortByBits(self, arr: List[int]) -> List[int]:
        bits = [[] for _ in range(32)]
        for num in arr:
            ones = 0
            for i in bin(num)[2:]:
                if i == '1':
                    ones += 1
            bits[ones].append(num)

        sortedBits = []
        for bit in bits:
            bit.sort()
            sortedBits.extend(bit)

        return sortedBits


def main():
    sol = Solution()
    print(sol.sortByBits([0,1,2,3,4,5,6,7,8]))
    print(sol.sortByBits([1024,512,256,128,64,32,16,8,4,2,1]))

main()
```

##LC #1022
Status: solved

```python
from typing import Optional
# Definition for a binary tree node.
class TreeNode:
  def __init__(self, val=0, left=None, right=None):
    self.val = val
    self.left = left
    self.right = right

class Solution:
  # Time: 35 m
  # Runtime: O(V*E)
  # Space: O(V) for stack
  def sumRootToLeaf(self, root: Optional[TreeNode]) -> int:
    total = 0
    bit = 0
    stack = [root]
    while stack:
      next = stack.pop()
      if next == -1:
        bit = bit >> 1
        continue
      bit = (bit << 1) | next.val
      stack.append(-1)
      if next.right:
        stack.append(next.right)
      if next.left:
        stack.append(next.left)
      if not next.right and not next.left:
        total += bit
    return total


def main():
  sol = Solution()
  print(sol.sumRootToLeaf())

```

## LC #868
Status: Solved

```python
class Solution:
  # Time: 12 m
  # Runtime: O(logn)
  # Space: O(1)
  def binaryGap(self, n: int) -> int:
    maxGap, curGap = 0, 0
    # find first 1
    while n & 1 != 1:
      n = n >> 1
    while n > 0:
      if n & 1 == 1:
        maxGap = max(maxGap, curGap)
        curGap = 0
      n = n >> 1
      curGap += 1
    return maxGap

def main():
  sol = Solution()
  print(sol.binaryGap(22))
  print(sol.binaryGap(8))
  print(sol.binaryGap(5))

  print(sol.binaryGap(3))

main()
```

##LC #1461

```python
class Solution:
  # Time: 23 m
  # Runtime: O(n*k)
  # Space: O(2^k)
  def hasAllCodes(self, s: str, k: int) -> bool:
    codes = set()
    for i in range(len(s) - k + 1):
      if s[i:i+k] not in codes:
        codes.add(s[i:i+k])
    codeCount = 2
    for i in range(1, k):
      codeCount *= 2

    return codeCount == len(codes)


def main():
  sol = Solution()
  print(sol.hasAllCodes("00110110", 2))
  print(sol.hasAllCodes("0110", 1))
  print(sol.hasAllCodes("0110", 2))


main()
```

## LC #696
Status: Solved

```python
class Solution:
  # Time: 18 m
  # Runtime: O(n)
  # Space: O(1)
  def countBinarySubstrings(self, s: str) -> int:
    count = 0
    prevConseq, curConseq = 0, 1
    for i in range(1, len(s) + 1):
      if i >= len(s) or s[i - 1] != s[i]:
        count += min(prevConseq, curConseq)
        prevConseq = curConseq
        curConseq = 0
      curConseq += 1
    return count

def main():
  sol = Solution()
  print(sol.countBinarySubstrings("00110011"))
  print(sol.countBinarySubstrings("10101"))

  print(sol.countBinarySubstrings("10"))
  print(sol.countBinarySubstrings("1"))
  print(sol.countBinarySubstrings("101"))

main()
```

## LC #693
Status: Solved

```python
class Solution:
  # 18 m
  # Runtime Complexity: O(len(n))
  # Space Complexity: O(1)
  def hasAlternatingBits(self, n: int) -> bool:
    alternating = True
    # get the rightmost bit
    bit = n & 1
    n = n >> 1
    while n and alternating:
      curRight = n & 1
      alternating = bit ^ curRight
      n = n >> 1
      bit = curRight
    return bool(alternating)

def main():
  sol = Solution()
  print(sol.hasAlternatingBits(1))

main()
```

## LC #190
Status: Acceptable, but runtime & space are slow, so there's a more elegant solution

```python
class Solution:
  # Time: 19 m
  # Runtime Complexity: O(32) since we traverse at most 32 elements
  # Space Complexity O(32) since we create a new digit
  def reverseBits(self, n: int) -> int:
    reversed = ['0'] * 34
    reversed[1] = 'b'
    orig = str(bin(n))
    for i in range(len(orig) - 2):
      reversed[len(orig) - 1 - i] = str(orig[2 + i])
    return int("".join(reversed), 2)

def main():
  sol = Solution()
  print(sol.reverseBits(43261596))
  print(sol.reverseBits(964176192))
  print(sol.reverseBits(2147483644))

main()
```

## LC #401

```python
class Solution:
    def readBinaryWatchRec(self, hours, minutes, turnedOn, curHours, curMinutes, times, curI, curJ):
            if curHours > 11 or curMinutes > 59 or turnedOn < 0:
                return
            if turnedOn == 0:
                if curHours < 12 and curMinutes < 60:
                    if curMinutes < 10:
                        timeStr = str(curHours) + ':0' + str(curMinutes)
                    else:
                        timeStr = str(curHours) + ':' + str(curMinutes)
                    times.add(timeStr)
            else:
                for i in range(curI, len(hours)):
                    self.readBinaryWatchRec(hours, minutes, turnedOn - 1, curHours + hours[i], curMinutes, times, i + 1, curJ)
                for j in range(curJ, len(minutes)):
                    self.readBinaryWatchRec(hours, minutes, turnedOn - 1, curHours, curMinutes + minutes[j], times, curI, j + 1)

    # time: 35 m
    # Runtime Complexity: O(turnedOn * minutes * hours)
    # Space Complexity: O(minutes * hours)
    def readBinaryWatch(self, turnedOn: int) -> List[str]:
        hours = [8,4,2,1]
        minutes = [32,16,8,4,2,1]
        times = set()

        self.readBinaryWatchRec(hours, minutes, turnedOn, 0, 0, times, 0, 0)
        return list(times)

```

## LC #67

Status: Solved, but lengthy and ugly. Recommend revisiting

```python
class Solution:
  # 49 m
  # Time Complexity: max(len(a), len(b))
  # Space Complexity: max(len(a), len(b))
  def addBinary(self, a: str, b: str) -> str:
    addition = []
    shorter, longer = (a, b) if len(a) <= len(b) else (b, a)
    carry = 0
    for i in range(min(len(a), len(b))):
      if a[-1-i] != b[-1-i]:
        if not carry:
          addition.append('1')
        else:
          addition.append('0')
      else:
        addition.append(str(carry))
        carry = 0
        if a[-1-i] == '1':
          carry = 1

    for j in range(len(shorter), len(longer)):
      if longer[-1-j] == '1' and carry:
        addition.append('0')
      elif longer[-1-j] == '0' and carry:
        addition.append(str(carry))
        carry = 0
      else:
        addition.append(longer[-1-j])

    if carry:
      addition.append(str(carry))


    return ''.join(addition[::-1])


def main():
  sol = Solution()
  print(sol.addBinary("11", "1"))
  print(sol.addBinary("11", "11"))

  print(sol.addBinary("1010", "1011"))

  print(sol.addBinary("1", "1"))
  print(sol.addBinary("111", "111"))
  print(sol.addBinary("100", "111"))

  print(sol.addBinary("100", "110010"))

  print(sol.addBinary("110010", "10111"))

main()
```
